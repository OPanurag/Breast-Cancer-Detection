```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
print(cancer)
```

```python
df = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns=np.append(cancer['feature_names'], ['target']))
df.head()
```

```python
df.tail()
```

```python
plt.figure(figsize=(25,10))
sns.pairplot(df, hue='target', vars=['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'])
```

```python
sns.countplot(x='target', data=df)
```

```python
plt.figure(figsize = (25,10))
sns.heatmap(df.corr(), annot=True, linewidth=0.5)
```

```python
x = df.drop(['target'], axis=1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test = tts(x, y, test_size=0.25, random_state=10)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
model = SVC()
```

```python
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
pred
```

```python
rep = classification_report(y_test, pred)
print(rep)
```

```python
mat = confusion_matrix(y_test, pred)
print(mat)
```

```python
sns.heatmap(mat, annot=True)
```
